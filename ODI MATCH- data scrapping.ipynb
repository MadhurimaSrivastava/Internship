```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
odi= requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
odi
```

```python
soup= BeautifulSoup(odi.content,'html.parser')
soup
print(soup.prettify())
```

```python
oditeam= []


for i in soup.find_all('span',class_="u-hide-phablet"):
    oditeam.append(i.text)

ot=oditeam[0:10]
ot
```

```python
first_team_match= soup.find('td',class_="rankings-block__banner--matches")
first_team_match.text
```

```python
ma=[first_team_match.text]
ma
```

```python
first_match_points = soup.find('td',class_="rankings-block__banner--points")
first_match_points.text

pnt=[first_match_points.text]
pnt
```

```python
first_match_ranking= soup.find('td',class_="rankings-block__banner--rating u-text-right")
first_match_ranking.text.replace('\n',"")

rk=[first_match_ranking.text.replace('\n',"")]
rk
```

```python
Odi_matches=[]
odi_points=[]


for match in soup.find_all('td',class_="table-body__cell u-center-text"):
    Odi_matches.append(match.text)
    
for t in soup.find_all('td',class_="table-body__cell u-center-text"):
    odi_points.append(t.text)


    
Odi_matches
odi_points
```

```python
matches= Odi_matches+ma
matches
match=matches[0:10]
match
```

```python
points= odi_points + pnt
points

pp = points[0:10]
pp
```

```python
odi_rating=[]

for r in soup.find_all('td',class_="table-body__cell u-text-right rating"):
    odi_rating.append(r.text)

odi_rating
```

```python
rating= odi_rating + rk
rating

ratings= rating[0:10]
ratings
```

```python
df = pd.DataFrame({"Teams":ot,"Matches":match,"points":pp,"Rating":ratings})
df
```
